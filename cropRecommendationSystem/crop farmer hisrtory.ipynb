```python
import random
crop_data = {
    "Banana": {
        "time_taken": range(11, 16),
        "sowing_month": range(1, 13),
        "expected_harvest": 60
    },
    "Blackgram": {
        "time_taken": range(2, 4),
        "sowing_month": range(6, 8),
        "expected_harvest": 930
    },
    "Chickpea": {
        "time_taken": range(4, 6),
        "sowing_month": range(10, 12),
        "expected_harvest": 12
    },
    "Coconut": {
        "time_taken": range(36, 48),
        "sowing_month": range(1, 13),
        "expected_harvest": 9000
    },
    "Coffee": {
        "time_taken": range(24, 30),
        "sowing_month": range(6, 8),
        "expected_harvest": 3000
    },
    "Cotton": {
        "time_taken": range(5, 6),
        "sowing_month": range(5, 7),
        "expected_harvest": 500
    },
    "Grape": {
        "time_taken": range(5, 7),
        "sowing_month": range(1, 13),
        "expected_harvest": 50
    },
    "Jute": {
        "time_taken": range(8, 10),
        "sowing_month": range(5, 7),
        "expected_harvest": 2.5
    },
    "Kidneybeans": {
        "time_taken": range(3, 4),
        "sowing_month": range(7, 9),
        "expected_harvest": 12
    },
    "Lentil": {
        "time_taken": range(3, 4),
        "sowing_month": range(10, 12),
        "expected_harvest": 400
    },
    "Maize": {
        "time_taken": range(3, 4),
        "sowing_month": range(6, 8),
        "expected_harvest": 3143
    },
    "Mango": {
        "time_taken": range(5, 8),
        "sowing_month": range(1, 13),
        "expected_harvest": 16
    },
    "Mothbeans": {
        "time_taken": range(3, 4),
        "sowing_month": range(7, 9),
        "expected_harvest": 140
    },
    "Mungbeans": {
        "time_taken": range(2, 3),
        "sowing_month": range(6, 8),
        "expected_harvest": 6
    },
    "Muskmelon": {
        "time_taken": range(2, 3),
        "sowing_month": range(2, 4),
        "expected_harvest": 150
    },
    "Orange": {
        "time_taken": range(8, 12),
        "sowing_month": range(1, 13),
        "expected_harvest": 9
    },
    "Papaya": {
        "time_taken": range(9, 12),
        "sowing_month": range(1, 13),
        "expected_harvest": 30
    },
    "Pigeonpeas": {
        "time_taken": range(5, 7),
        "sowing_month": range(1, 13),
        "expected_harvest": 25
    },
    "Pomegranate": {
        "time_taken": range(10, 12),
        "sowing_month": range(1, 13),
        "expected_harvest": 11
    },
    "Rice": {
        "time_taken": range(4, 6),
        "sowing_month": range(6, 8),
        "expected_harvest": 2000
    },
    "Watermelon": {
        "time_taken": range(2, 3),
        "sowing_month": range(1, 13),
        "expected_harvest": 30000
    }
}


num_farmers = 15
num_records = 15
def random_nearby_dynamic(num,mag=False):
    magnitude = len(str(abs(int(num))))
    deviation = magnitude / mag if mag else 10
    return num + random.uniform(-deviation, deviation)


data = {}
crops = list(crop_data.keys())
n = len(crops)

def find_crop(month):
    while True:
        temp = random.randint(1,n)
        crop = crop_data[crops[temp-1]]
        #print(month,crop["sowing_month"])
        if month in crop["sowing_month"]:
            expected = random_nearby_dynamic(crop["expected_harvest"])
            return temp,random.randint(crop["time_taken"][0],crop["time_taken"][-1]+1),expected,expected-expected*random.random()



for i in range(num_farmers):
    data[i] = []
    start_year=2001
    start_month=6
    for j in range(num_records):
        record = {"crop":"" , "sowing_month":"","sowing_year":"","harvest_month":"","harvest_year":"","expected":"","actual":""}
        print(start_month)
        key, time_taken , expected , observed = find_crop(start_month)
        record["crop"] = key
        record["expected"] = expected
        record["actual"] = observed
        record["sowing_month"] = start_month
        record["sowing_year"] = start_year
        start_month+=time_taken
        if start_month>12:
            start_year+=int(start_month/12)
            start_month%=12
            if start_month == 0: start_month = 1
        record["harvest_month"] = start_month
        record['harvest_year'] = start_year

        data[i].append(record)
    
print(data)


```
